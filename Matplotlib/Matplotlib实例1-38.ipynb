# 散点图(Scatter plot)

```python
# %matplotlib widget
# 导入公用模块
import numpy as np
import matplotlib.pyplot as plt

N = 50

x = np.linspace(0., 10., N)
y = np.sin(x)**2 + np.cos(x)
```

变量x是0到10的50个数据的数组，变量y是sin(x)和cos(x)的平方之和。可以用以下代码以散点图的形式可视化x轴上的变量x和y轴上的变量y

```python
plt.figure()
plt.scatter(x, y)
```

## 为了使其美观，可以减少每个数据大小并给标签添加下面的代码

```python
plt.scatter(x, y, s=15, label=r'$y = sin^2(x)+cos(x)$')
```

```python
color = 'r'
```

```python
plt.axis('equal')
```

## 完整代码

```python
N = 50

x = np.linspace(0., 10., N)
y = np.sin(x)**2 + np.cos(x)

plt.figure()
plt.scatter(x, y, s=15, label=r'$ y  = sin^2(x) + cos(x)$', color='r')
plt.axis('equal')

plt.legend()
plt.xlabel(r'$x$ (red)', color='r')
plt.ylabel(r'y')

# plt.savefig('file/scatter2.png', dpi=300, bbox_inches='tight', facecolor='w')
```

## 更改标记样式

>marker = 'x'

```python

N = 50

x = np.linspace(0., 10., N)
y = np.sin(x)**2 + np.cos(x)

plt.figure()
plt.scatter(x, y, s=15, label=r'$ y  = sin^2(x) + cos(x)$',
            color='r', marker='x')
plt.axis('equal')

plt.legend()
plt.xlabel(r'x', color='r')
plt.ylabel(r'y')
```

### 将字母作为标记样式

>这里使用希腊符号 beta 作为标记样式。也可以使用其他字母来更改它，例如  a，B，C，d 或 ** 1、2、3** 等。

```python
np.random.seed(100)

N = 50

randx = np.random.random(N) * 100
randy = np.random.random(N) * 100
plt.figure(figsize=(7, 6))

plt.scatter(randx, randy, marker=r'$\beta$', s=150, color='darkorange')

plt.axis('equal')

plt.xlabel('randx')
plt.ylabel('randy')

plt.tight_layout()
```
