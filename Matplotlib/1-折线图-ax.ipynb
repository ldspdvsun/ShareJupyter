# 折线图-ax

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
%matplotlib inline
````

````python
fig, ax = plt.subplots(figsize=(14, 7))
A = np.arange(1, 6)
B = A**2
ax.plot(A, B)
````

## 线条样式
>linestyle='-.'

字符|类型|字符|类型
---|---|---|---
`'-'`|实线|`'--'`|虚线
`'-.'`|虚点线|`':'`|点线
`'None'`|空白|`' '`|空白
`'solid'`|实线|`'dashed'`|虚线
`'dashdot'`|虚线点|`'dotted'`|点线

ax.plot(A,B,linestyle='dashdot')

linestyle=['-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted']

````python
fig, ax = plt.subplots(figsize=(14, 7))
A = np.arange(1, 6)
B = A**2
ax.plot(A, B, linestyle='dashdot')
````

## 数据点属性
### 数据点样式
>设置数据点样式：marker='o'

字符|类型|字符|类型
---|---|---|---
`','`|像素点|`'_'`|横线点
`'o'`|圆点|`'+'`|加号点
`'<'`|左三角点|`'v`|下三角点
`'>'`|右三角点|`'1'`|下三叉点
`'2'`|上三叉点|`'3'`|左三叉点
`'4'`|右三叉点|`'s'`|正方点
`'p'`|五角点|`'*'`|星形点
`'h'`|六边形点1|`'H'`|六边形点2
`'x'`|瘦乘号点|`'X'`|实心乘号点
`'d'`|瘦菱形点|`'D'`|实心菱形点

### 数据点颜色
>设置数据点颜色：markerfacecolor='r'

### 数据点大小
>设置数据点大小：markersize=10

ax.plot(A,B,linestyle='dashdot',color='c',marker='>',markerfacecolor='r',markersize=20)

````python
fig, ax = plt.subplots(figsize=(14, 7))
A = np.arange(1, 6)
B = A**2
ax.plot(A,
        B,
        linestyle='dashdot',
        color='c',
        marker='>',
        markerfacecolor='r',
        markersize=20)
````

## 线条宽度

ax.plot(A,B,linestyle='dashdot',color='c',marker='>',markerfacecolor='r',markersize=20,linewidth=3)

````python
fig, ax = plt.subplots(figsize=(14, 7))
A = np.arange(1, 6)
B = A**2
ax.plot(A,
        B,
        linestyle='dashdot',
        color='c',
        marker='>',
        markerfacecolor='r',
        markersize=20,
        linewidth=3)
````

## 设置线条透明度
>alpha=0.5

ax.plot(A,B,linestyle='dashdot',color='c',marker='>',markerfacecolor='r',markersize=20,linewidth=3,alpha=0.5)

````python
fig, ax = plt.subplots(figsize=(14, 7))
A = np.arange(1, 6)
B = A**2
ax.plot(A,
        B,
        linestyle='dashdot',
        color='c',
        marker='>',
        markerfacecolor='r',
        markersize=20,
        linewidth=3,
        alpha=0.5)
````

## 背景格
>ax.grid(which='both',axis='both')

设置背景格：which='major', 'minor', 'both' ；axis='x', 'y', 'both'

````python
fig, ax = plt.subplots(figsize=(14, 7))
A = np.arange(1, 6)
B = A**2
ax.plot(A,
        B,
        linestyle='dashdot',
        color='c',
        marker='>',
        markerfacecolor='r',
        markersize=20,
        linewidth=3,
        alpha=0.5)
ax.grid(which='both', axis='both')
````

## X轴坐标值旋转
>ax.xaxis.set_tick_params(rotation=45)

````python
fig, ax = plt.subplots(figsize=(14, 7))
A = np.arange(1, 6)
B = A**2
ax.plot(A,
        B,
        linestyle='dashdot',
        color='c',
        marker='>',
        markerfacecolor='r',
        markersize=20,
        linewidth=3,
        alpha=0.5)
ax.grid(which='both', axis='both')
# 旋转45度
ax.xaxis.set_tick_params(rotation=45)
````

## 设置X,Y轴中间刻度及网格线

>设置xy坐标轴每个间隔之间显示刻度以及网格线（两个刻度之间自动分配为5份）：ax.minorticks_on()  

````python
fig, ax = plt.subplots(figsize=(14, 7))
A = np.arange(1, 6)
B = A**2
ax.plot(A,
        B,
        linestyle='dashdot',
        color='c',
        marker='>',
        markerfacecolor='r',
        markersize=20,
        linewidth=3,
        alpha=0.5)
ax.grid(which='both', axis='both')
# 旋转45度
ax.xaxis.set_tick_params(rotation=45)
# 设置xy坐标轴每个间隔之间显示刻度以及网格线
ax.minorticks_on()
````

## 重新设置x，y轴坐标值范围
重新设置X,Y轴坐标值后，相当于重新展示原有数据图的对应坐标值部分

>设置x轴坐标值范围，按照刻度值重新编译：ax.set_xlim(0, 10) 

>设置y轴坐标值范围，按照刻度值重新编译：ax.set_ylim(10, 36) 

````python
fig, ax = plt.subplots(figsize=(14, 7))
A = np.arange(1, 6)
B = A**2
ax.plot(A,
        B,
        linestyle='dashdot',
        color='c',
        marker='>',
        markerfacecolor='r',
        markersize=20,
        linewidth=3,
        alpha=0.5)
ax.grid(which='both', axis='both')
# 旋转45度
ax.xaxis.set_tick_params(rotation=45)
# 设置xy坐标轴每个间隔之间显示刻度以及网格线
ax.minorticks_on()
# 设置x轴坐标值范围，按照刻度值重新编译
ax.set_xlim(0, 10)
# 设置y轴坐标值范围，按照刻度值重新编译
ax.set_ylim(10, 36)
````

## 设置X,Y坐标轴刻度值的位置
>设置Y轴标签调整到右侧：ax.yaxis.tick_right()

````python
fig, ax = plt.subplots(figsize=(14, 7))
A = np.arange(1, 6)
B = A**2
ax.plot(A,
        B,
        linestyle='dashdot',
        color='c',
        marker='>',
        markerfacecolor='r',
        markersize=20,
        linewidth=3,
        alpha=0.5)
ax.grid(which='both', axis='both')
# 旋转45度
ax.xaxis.set_tick_params(rotation=45)
# 设置xy坐标轴每个间隔之间显示刻度以及网格线
ax.minorticks_on()
# 设置x轴坐标值范围，按照刻度值重新编译
ax.set_xlim(0, 10)
# 设置y轴坐标值范围，按照刻度值重新编译
ax.set_ylim(10, 36)
# 设置Y轴标签调整到右侧
ax.yaxis.tick_right()
ax.xaxis.tick_top()
````

## 设置X轴坐标值间隔
```py
start, end = ax.get_xlim()
print(start)
print(end)
ax.xaxis.set_ticks(np.arange(start, end, 1))
```

````python
fig, ax = plt.subplots(figsize=(14, 7))
A = np.arange(1, 6)
B = A**2
ax.plot(A,
        B,
        linestyle='dashdot',
        color='c',
        marker='>',
        markerfacecolor='r',
        markersize=20,
        linewidth=3,
        alpha=0.5)
ax.grid(which='both', axis='both')
# 旋转45度
ax.xaxis.set_tick_params(rotation=45)
# 设置xy坐标轴每个间隔之间显示刻度以及网格线
ax.minorticks_on()

# 设置x轴坐标值范围，按照刻度值重新编译
ax.set_xlim(0, 10)
# 设置y轴坐标值范围，按照刻度值重新编译
ax.set_ylim(0, 36)

start, end = ax.get_xlim()
print(start)
print(end)
ax.xaxis.set_ticks(np.arange(start, end, 0.5))
````

## 相关注释
### X,Y轴标签，标题
* ax.set_xlabel('X轴标签')
* ax.set_ylabel('Y轴标签')
* ax.set_title('标题')
### 某个位置上添加注释
ax.text(x轴坐标,y轴坐标,'注释内容')
### 设置图例
ax.legend(['标签列表'],loc='位置')
### 为某个数据点添加注释，箭头
ax.annotate('注释内容',xy=(箭头指向位置横坐标,箭头指向位置纵坐标),xytext=(箭头末尾注释位置横坐标,箭头末尾注释位置纵坐标),arrowprops=dict(facecolor='箭头颜色',shrink=箭头缩放比（值越小箭头越长）,headlength=箭头三角部分长度,headwidth=箭头三角部分宽度))

````python
fig, ax = plt.subplots(figsize=(14, 7))
A = np.arange(1, 6)
B = A**2
ax.plot(A,
        B,
        linestyle='dashdot',
        color='c',
        marker='>',
        markerfacecolor='r',
        markersize=20,
        linewidth=3,
        alpha=0.5)
ax.grid(which='both', axis='both')

ax.set_xlabel('xlabel')
ax.set_ylabel('ylabel')
ax.set_title('A-B')

ax.text(4, 2, '4,2-zhushi')

ax.legend(['X-A_Y-B'], loc='upper center')

ax.annotate('zhushi',
            xy=(2, 4),
            xytext=(2.5, 5),
            arrowprops=dict(facecolor='black',
                            shrink=0.05,
                            headlength=30,
                            headwidth=30))

# 旋转45度
ax.xaxis.set_tick_params(rotation=45)
````

# 多个折线

````python
fig, ax = plt.subplots(figsize=(14, 7))
A = np.arange(1, 6)
B = A**2
ax.plot(A, A, linestyle='-', color='r', marker='_')
ax.plot(A, B, linestyle='--', color='g', marker='x')
ax.plot(B, A, linestyle='-.', color='b', marker='+')
ax.plot(B, B, linestyle=':', color='y', marker='v')

ax.grid(which='both', axis='both')

ax.set_xlabel('xlabel')
ax.set_ylabel('ylabel')
ax.set_title('A-B')

ax.text(4, 2, '4,2-zhushi')

ax.legend(['AA', 'AB', 'BA', 'BB'], loc='upper center')

ax.annotate('zhushi',
            xy=(5, 5),
            xytext=(15, 10),
            arrowprops=dict(facecolor='black',
                            shrink=0.05,
                            headlength=30,
                            headwidth=30))

# 旋转45度
ax.xaxis.set_tick_params(rotation=45)
````

```py
fig, axs = plt.subplots(2,1)
fig.suptitle('Vertically stacked subplots')
axs[0].plot(A, B)

```
(**2,1**)分别代表将要绘制的子图有**2行，1列**

````python
A = np.arange(1, 6)
B = A**2
fig, axs = plt.subplots(2, 1)
fig.suptitle('Vertically stacked subplots')
axs[0].plot(A, B, color='m')
axs[1].plot(B, A, color='c')
````
